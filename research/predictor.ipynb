```python
import pandas as pd
from sklearn import tree
import numpy as np
from sklearn import tree
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
import json
import joblib

data = pd.read_csv("Training.csv")
X = data.iloc[:,:-1]
y = data["prognosis"]
org = y
le = preprocessing.LabelEncoder()
le.fit(y)
y = le.transform(y)

features = list(X.columns[:132])
# print(features)

data.head(50)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.33,random_state=20)
model = RandomForestClassifier(n_estimators=10)
model.fit(X_train,y_train)
print(model.score(X_test,y_test))
```

```python
clf = RandomForestClassifier(n_estimators=10)
clf.fit(X,y)
```

```python
def askme(symptoms):
    print("I will ask you 10-15question about how are experiencing. Please stay with me...\n")
    j = 0
    for i in range(14):
        print("Ok " + str(i+1) + " one... Please enter from 1 to 10 to choose and 0 to say NO!\n")
        for k in range(j,min(j+10,len(features))):
            print(str(k+1) + " " + features[k], end=" ")
        j+= 10
        x = input()
        l = [int(m) for m in x.strip().split()]
        if x == 0:
            continue
        for m in range(len(l)):
            symptoms[l[m]] = 1
    return symptoms
            
```

```python
symptoms = [0]*132
symptoms = askme(symptoms)

val = clf.predict([symptoms])
# print(symptoms)
# idx = np.where(y==val)
# print("You have " + org[idx[0]] + ". You should consult doctor...")
disease = le.inverse_transform(val)
print("Predicted disease is " + disease[0])
# print(val.dtype)
# print(val.dtype)
```
